```python
fires_2017 = pd.read_csv("forest_fires_1.csv")
```

```python
fires_2017.drop(fires_2017.index[[43, 49, 50, 51]], inplace=True)
```

```python
fires_2017
```

```python
plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
})
```

```python
#color = plt.rcParams['axes.prop_cycle'].by_key()['color'][3]
ax = fires_2017.sort_values(by="Current year", ascending=False).head(5).plot(x='Country', y='Current year', kind='barh', color='#C30202', figsize=(10,5), legend=False)

for i, bar in enumerate(ax.patches):
    # print(i)
    value = str(int(round(bar.get_width()))) + " (ha)"
    x_pos = bar.get_width() - 1
    y_pos = i
    # ax.text(x_pos, y_pos, value, verticalalignment='center')
    ax.text(x_pos, y_pos, value, va='center', ha='right', color='black')

ax.set_title ('5 EU countries with biggest forest areas burnt in 2017', pad = 40)
ax.legend().set_visible(False)
ax.title.set_fontsize(30)
ax.title.set_fontname('Gill Sans MT')

plt.gcf().text(0.5, 0.92, "Hectares (ha) burnt, by country", ha='center', fontsize=15)

ax.set_xlabel("(ha) burnt")

ax.tick_params(which='major', left=False, bottom=False)

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['right'].set_visible(False)

ax.tick_params(which='major', left=False, bottom=True, grid_alpha=0.5)
ax.set_facecolor('#faebd7')

ax.grid(False, axis='x')
ax.set_axisbelow(False)
#ax.set_xlim(0, 10)
ax.yaxis.label.set_visible(False)
plt.show()
```

```python
burned_areas_by_year = pd.read_csv("forest_fires_3_1.csv")
```

```python
burned_areas_by_year.head()
```

```python
burned_areas_by_year.rename(columns={'Burnt Area (ha) ': 'Year'},
                            inplace=True)
```

```python
burned_areas_by_year.head()
```

```python
portugal_fire_hist_column_1= burned_areas_by_year['Portugal']
```

```python
years_port = burned_areas_by_year['Year']
```

```python
chronicle_of_portugal_burnt_areas  = pd.concat([years_port, portugal_fire_hist_column_1], axis = 1, sort=True, ignore_index=True)
```

```python
chronicle_of_portugal_burnt_areas.tail()
```

```python
chronicle_of_portugal_burnt_areas.rename(columns={0: 'Year', 
                                                  1: 'Hectares burnt',
                                                             }, 
                                                   inplace=True)
```

```python
ax = chronicle_of_portugal_burnt_areas.plot(x='Year', y='Hectares burnt', color='#C30202', figsize=(10,5), linewidth=0.8)

ax.set_title ('Forests lost in Portugal, 1980-2017', pad = 40)
ax.legend().set_visible(False)
ax.title.set_fontsize(30)
ax.title.set_fontname('Gill Sans MT')

ax.set_facecolor('#FBF2F2')

ax.set_xlabel("")

ax.set_xticks([1980, 1985, 1990, 1995, 2000, 2005, 2010, 2017])

plt.gcf().text(0.5, 0.92, "Hectares (ha) burnt by year", ha='center', fontsize=15)

plt.legend(bbox_to_anchor=(0, 0.98, 1., 0.), ncol=0.1,
           mode="normal")


ax.spines['top'].set_visible(False)
#ax.set_ylim(0, 461079)

plt.show()
```
